```python
def clean_text(contents):
    tmp = []
    for content in contents:
        if type(content) is type('1'):
            content = content.replace('\n','')
            content = re.sub(' +', ' ', content)
            tmp.append(content[1:-1])
        else:
            tmp.append(content)
    return tmp
```
